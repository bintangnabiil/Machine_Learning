```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Automobile.csv')

if 'symboling' not in df.columns:
  print("Kolom 'symboling' tidak ditemukan!")
else:
  df['symboling'] = pd.to_numeric(df['symboling'], errors='coerce')

x = df[['wheel-base', 'length', 'width', 'height', 'curb-weight']]
y = df['symboling']

x = x.dropna()
y = y[x.index]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)

print("Intercept (w0): ", model.intercept_)
print("Koefisien model: ", model.coef_)

y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("Mean Squared Error (MSE): ", mse)
print("Root Mean Squared Error (RMSE): ", rmse)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color="red", linestyle="--")
plt.xlabel("Actual Symboling")
plt.xlabel("Predicted Symboling")
plt.ylabel("Actual vs Predicted Values")
plt.show()
```

# Persamaan Matematika
## 1. Persamaan Model Linear Regression
Kode ini menggunakan Linear Regression yang merupakan model regresi berbasis persamaan garis linear. Kode ini mencoba memprediksi variabel target y (symboling) berdasarkan kombinasi linear dari beberapa variabel independen x (wheel-base, length, width, height, curb-weight). Untuk persamaannya adalah sebagai berikut:

$y = w_0 + w_1x_1 + w_2x_2 +...+ w_nx_n$

Di mana:
- $y$= nilai prediksi
- $w_0$ = intercept (bias)
- $w_1, w_2, ..., w_n$ = koefisien regresi yang menunjukkan pengaruh masing-masing fitur terhadap target
- $x_1, x_2, ..., x_n$ = nilai fitur input untuk prediksi

##2. Evaluasi Model
#a.) Mean Squeared Error (MSE)
MSE digunakan untuk mengukur rata-rata kesalahan kuadarat antara nilai aktual dan prediksi:

$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$

Di mana:
- $n$ = jumlah sampel
- $y_i$ = nilai target aktual untuk sampel ke-$i$
- $\hat{y}_i$ = nilai yang diprediksi oleh model untuk sampel ke-$i$

#b.) Root Mean Squared Error (RMSE)
RMSE digunakan untuk mengukur seberapa besar perbedaan nilai yang diprediksi oleh model dengan nilai aktual. RMSE dihitung sebagai akar dari MSE:
$RMSE = \sqrt{MSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$

Di mana:
- $n$ = jumlah sampel dalam dataset
- $y_i$ = nilai aktual pada sampel ke-$i$
- $\hat{y}_i$ = nilai prediksi dari model pada sampel ke-$i$
